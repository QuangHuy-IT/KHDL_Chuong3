```python
import pandas as pd

# Đọc dữ liệu từ file CSV của bạn
df = pd.read_csv("bds_clean.csv")

# Xem nhanh 5 dòng đầu để kiểm tra
print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df['Diện tích (m²)'], bins=30, edgecolor='black', color='orange')
plt.title('Phân bố diện tích các bất động sản')
plt.xlabel('Diện tích (m²)')
plt.ylabel('Số lượng')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Giá (tỷ)'], bins=30, edgecolor='black', color='gold')
plt.title('Phân bố giá bán (tỷ đồng)')
plt.xlabel('Giá (tỷ)')
plt.ylabel('Số lượng')
plt.show()
```

```python
plt.figure(figsize=(7,5))
df.groupby('Tình trạng nội thất')['Giá (tỷ)'].mean().plot(kind='bar', color=['skyblue', 'orange', 'green'])
plt.title('Giá trung bình theo tình trạng nội thất')
plt.xlabel('Tình trạng nội thất')
plt.ylabel('Giá trung bình (tỷ)')
plt.show()
```

```python
df['Ngày đăng'] = pd.to_datetime(df['Ngày đăng'], errors='coerce', format='%d/%m/%Y')

df_sorted = df.groupby('Ngày đăng')['Giá (tỷ)'].mean().reset_index()

plt.figure(figsize=(18,10))
plt.plot(df_sorted['Ngày đăng'], df_sorted['Giá (tỷ)'], 
         marker='o', linestyle='-', color='teal', linewidth=2)

plt.title('Biểu đồ đường: Giá trung bình nhà đất theo ngày đăng', fontsize=14, fontweight='bold')
plt.xlabel('Ngày đăng', fontsize=12)
plt.ylabel('Giá trung bình (tỷ đồng)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

for i, v in enumerate(df_sorted['Giá (tỷ)']):
    plt.text(df_sorted['Ngày đăng'].iloc[i], v + 0.5, f"{v:.1f}", ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```
